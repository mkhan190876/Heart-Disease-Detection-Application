```python
X = df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
logisticRegr = LogisticRegression()
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
logisticRegr.fit(x_train, y_train)
```

```python
predictions = logisticRegr.predict(x_test)
```

```python
accuracy = logisticRegr.score(x_test,y_test)
print("The accuracy of the Logistic Regression model is:", accuracy*100)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test,predictions)
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
clf_pred = clf.predict(x_test)
```

```python
accuracy = clf.score(x_test,y_test)
print("The accuracy of the Decision Tree model is:", accuracy*100)
```

```python
metrics.confusion_matrix(y_test,clf_pred)
```

```python
print(metrics.classification_report(y_test,clf_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
rf_pred = rf.predict(x_test)
print(rf_pred)
```

```python
accuracy = rf.score(x_test,y_test)
print("The accuracy of the Random Forest model is:", accuracy*100)
```

```python
metrics.confusion_matrix(y_test,rf_pred)
```

```python
print(metrics.classification_report(y_test,rf_pred))
```
